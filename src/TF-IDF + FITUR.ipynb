```python
import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler

import warnings
warnings.filterwarnings("ignore")
```

```python
data_train_ = pd.read_excel("../Data/data_train.xlsx").drop(['Unnamed: 0'],axis=1)
data_test_ = pd.read_excel("../Data/data_test.xlsx").drop(['Unnamed: 0'],axis=1)
```

```python
data_train_.head()
```

```python
data_test_.head()
```

```python
X_train = data_train_[['Tweet_Parsed','Following','Followers','Jumlah tweet']]
X_test = data_test_[['Tweet_Parsed','Following','Followers','Jumlah tweet']]
y_train = data_train_.drop(['Tweet','Tweet_Parsed','Following','Followers','Jumlah tweet'],axis=1)
y_test = data_test_.drop(['Tweet','Tweet_Parsed','Following','Followers','Jumlah tweet'],axis=1)
```

```python
ngram_range = (1,1)
min_df = 1
max_df = 1.
max_features = None
```

```python
tfidf = TfidfVectorizer(encoding='utf-8',
                        ngram_range=ngram_range,
                        stop_words=None,
                        lowercase=False,
                        max_df=max_df,
                        min_df=min_df,
                        max_features=max_features,
                        norm='l2')
```

```python
features_train = tfidf.fit_transform(X_train['Tweet_Parsed']).toarray()
labels_train = y_train
print(features_train.shape)
```

```python
features_test = tfidf.transform(X_test['Tweet_Parsed']).toarray()
labels_test = y_test
print(features_test.shape)
```

```python
fitur = tfidf.get_feature_names()
target = data_train_.drop(['Tweet','Tweet_Parsed','Following','Followers','Jumlah tweet'],axis = 1).columns
```

```python
data_train_tfidf = pd.DataFrame(data=features_train,columns=fitur)
data_train_tfidf = data_train_tfidf.join(X_train[['Following','Followers','Jumlah tweet']]).join(pd.DataFrame(data=y_train,columns=target))
data_train_tfidf.head()
```

```python
data_train_tfidf.loc[data_train_tfidf['Temp'] == 'Melankolis', ['Temp']] = 0
data_train_tfidf.loc[data_train_tfidf['Temp'] == 'Sanguinis', ['Temp']] = 1
data_train_tfidf.loc[data_train_tfidf['Temp'] == 'Koleris', ['Temp']] = 2
data_train_tfidf.loc[data_train_tfidf['Temp'] == 'Plegmatis', ['Temp']] = 3
data_train_tfidf.loc[data_train_tfidf['Temp'] == 'Plemgatis', ['Temp']] = 3
```

```python
data_train_tfidf.loc[data_train_tfidf['K'] == 'Guardian', ['K']] = 0
data_train_tfidf.loc[data_train_tfidf['K'] == 'Analyst', ['K']] = 1
data_train_tfidf.loc[data_train_tfidf['K'] == 'Idealist', ['K']] = 2
data_train_tfidf.loc[data_train_tfidf['K'] == 'Artisan', ['K']] = 3
```

```python
data_test_tfidf = pd.DataFrame(data=features_test,columns=fitur)
data_test_tfidf = data_test_tfidf.join(X_test[['Following','Followers','Jumlah tweet']]).join(pd.DataFrame(data=y_test,columns=target))
data_test_tfidf.head()
```

```python
data_test_tfidf.loc[data_test_tfidf['Temp'] == 'Melankolis', ['Temp']] = 0
data_test_tfidf.loc[data_test_tfidf['Temp'] == 'Sanguinis', ['Temp']] = 1
data_test_tfidf.loc[data_test_tfidf['Temp'] == 'Koleris', ['Temp']] = 2
data_test_tfidf.loc[data_test_tfidf['Temp'] == 'Plegmatis', ['Temp']] = 3
data_test_tfidf.loc[data_test_tfidf['Temp'] == 'Plemgatis', ['Temp']] = 3
```

```python
data_test_tfidf.loc[data_test_tfidf['K'] == 'Guardian', ['K']] = 0
data_test_tfidf.loc[data_test_tfidf['K'] == 'Analyst', ['K']] = 1
data_test_tfidf.loc[data_test_tfidf['K'] == 'Idealist', ['K']] = 2
data_test_tfidf.loc[data_test_tfidf['K'] == 'Artisan', ['K']] = 3
```

```python
def generate_ngrams(s):
    
    # Break sentence in the token, remove empty tokens
    tokens = [token for token in s.str.split(" ") if token != ""]
    
    # Use the zip function to help us generate n-grams
    # Concatentate the tokens into ngrams and return
    ngrams = np.concatenate(tokens)
    return ngrams
```

```python
tokensTest = generate_ngrams(data_test_['Tweet_Parsed']).tolist()
# tokensTrain = np.unique(tokensTrain).tolist()
tokensTest
```

```python
outOfVocab = []
for i in tokensTest:
    if i not in fitur:
        outOfVocab.append(i)
np.unique(outOfVocab).tolist()
```

```python
print("# out of vocabulary : ", len(outOfVocab))
print("# out of vocabulary (unique words) : ", len(np.unique(outOfVocab)))
```
